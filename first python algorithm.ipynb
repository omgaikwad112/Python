# THE MATKA KING 

```python
from collections import Counter
x=map(int,input().split())
matka=Counter(x)
bingo=list(Counter(matka).items())
a=sorted(list(Counter(matka).values()))[0]
print("THE LEAST FREQUENCY IS:-",a)
print("THE LEAST FREQUENT NUMBER'S ARE:-")
for i in bingo:
    if i[1]==a:
        print(i[0])
```
